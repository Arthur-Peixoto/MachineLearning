```python
# prompt: faça isso aqui dar certo: sb.pairplot(df,kind = 'kde')

import pandas as pd
from sklearn.impute import SimpleImputer
import numpy as np
import seaborn as sb # import seaborn

df = pd.read_csv('fake_data_2.csv')
df.head()
dfteste = df.copy()
dfteste.fillna(0, inplace=True)
print(dfteste)

imputer = SimpleImputer(missing_values=0, strategy='mean')
dfteste = df.copy()

q1 = dfteste['idade'].quantile(0.25)
q3 = dfteste['idade'].quantile(0.75)
iqr = q3 - q1
lLim = q1 - (1.5 * iqr)
hLim = q3 + (1.5 * iqr)
print(lLim)
print(hLim)
dfteste.loc[dfteste['idade'] < lLim, 'idade'] = lLim
print()
dfteste.loc[dfteste['idade'] > hLim, 'idade'] = hLim
print(dfteste)
dfteste['idade'].plot(kind = 'box')
sb.pairplot(df,kind = 'kde') # now it will work because seaborn is imported
```

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(strategy='mean')  # Replace with your desired strategy
df[['idade', 'salario', 'bonus']] = imputer.fit_transform(df[['idade', 'salario', 'bonus']])

# Now apply Normalizer
from sklearn.preprocessing import Normalizer
df_normalize = df.copy()
df_normalize[['idade', 'salario', 'bonus']] = Normalizer().fit_transform(df[['idade', 'salario', 'bonus']])
print(df_normalize)
```

```python
from sklearn.preprocessing import MinMaxScaler

df[['idade', 'salario', 'bonus']] = MinMaxScaler().fit_transform(df[['idade', 'salario', 'bonus']])

print(df)
```
